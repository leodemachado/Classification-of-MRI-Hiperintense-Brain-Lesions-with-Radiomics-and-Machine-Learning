```python
import numpy as np
import pandas as pd

import pycaret as pc
from pycaret.classification import *
```

```python
# read and prepare the dataframe:

miccai_3dt1 = pd.read_csv('MICCAI_3DT1.csv')
del miccai_3dt1['Unnamed: 0']

wmh_3dt1 = pd.read_csv('WMH_3DT1.csv')
del wmh_3dt1['Unnamed: 0']

# AChange binary classification for label 0 = scleroses, 1 = wmh;
def change(x):
    if x == 0.0: x = 'scleroses'
    else: x = 'cvl'
    return x

wmh_3dt1['class'] = wmh_3dt1['class'].map(change)
miccai_3dt1['class'] = miccai_3dt1['class'].map(change)

lesions = pd.DataFrame()
lesions = lesions.append(miccai_3dt1, ignore_index=True)
lesions = lesions.append(wmh_3dt1, ignore_index=True)
del lesions['Patient_ID']
```

```python
clfs = setup(lesions, target='class')
```

```python
compare_models()
```

```python
lr = create_model('lr')
```

```python
tuned = tune_model(lr)
```

```python
plot_model(tuned)
```

```python
plot_model(tuned, plot='feature')
```

```python
predict_model(tuned)
```
